# Questão 1
Uma doença se espalha de forma exponencial, com uma taxa de contágio de 0,05 por dia. Se inicialmente 1000 pessoas são infectadas, faça um gráfico que mostre como o número de pessoas infectadas varia com o tempo e determine quantas pessoas estarão infectadas após 30 dias. Utilize a equação $I = I_0 * e^{rt}$, onde $I_0$ é o número inicial de infectados, `r` é a taxa de contágio diária (em decimal) e $t$ é o tempo decorrido em dias.



```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = np.linspace(0,30,256, endpoint = True)
y = 1000 * np.exp(0.05*x)
fig = plt.figure(figsize = (10,10))
plt.plot(x, y, '-r', label=r'$I = {1000*e^{rt}}$')
plt.plot(x, y, '-r')
plt.grid()
plt.xlabel("Dias")
plt.ylabel("Pessoas Infectadas")
plt.legend(loc='upper left',prop={'size': 25})
plt.show()

result = 1000 * np.exp(0.05*30)
out = f'Pessoas  \space infectadas \space após \space 30 \space dias = {result:.0f}'
display(Math(out))
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x) = 3x^2 - 6^x +48$

2. $f(x) = \log_{10}(x^2 - 3x) - \log_{10}(x - 3)$

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = np.linspace(0,10,256, endpoint = True)
y = ((3*x**2) - (6**x) + 48)
fig = plt.figure(figsize = (8,8))
plt.plot(x, y, '-b', label=r'$f(x) = {3x^2-6^x+48}$')
plt.plot(x, y, '-b')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='lower left',prop={'size': 15})
plt.plot(x, y, '-b', label=r'$Descrescente')
plt.legend(loc='lower left',prop={'size': 15})
plt.show()

x = np.linspace(-1.5,4.5,256, endpoint = True)
y1 = x**2 - 3*x
y2 = x-3
y = np.log10(y1)-np.log10(y2)
fig = plt.figure(figsize = (8,8))
plt.plot(x, y, '-g', label=r'$f(x) = \log_{10}(x^2 - 3x) - \log_{10}(x - 3)$')
plt.plot(x, y, '-g')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='lower left',prop={'size': 15})
plt.plot(x, y, '-g', label=r'$Crescente')
plt.legend(loc='lower left',prop={'size': 15})
plt.show()
```
