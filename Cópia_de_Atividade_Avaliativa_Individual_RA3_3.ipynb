# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
#Renan Antonio Hammerschmidt Krefta

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')

f = x*(sp.exp(-x**2))
g = x+1

a, b = 0, 2

integral_value_f = float(sp.integrate(f, (x, a, b)))
integral_value_g = float(sp.integrate(g, (x, a, b)))

f = sp.lambdify(x, f, "numpy")
g = sp.lambdify(x, g, "numpy")

x_vals = np.linspace(-5, 5, 1000)

f_vals = f(x_vals)
g_vals = g(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label=r'$f(x) = x e^{-x^2}$')
plt.plot(x_vals, g_vals, label='$g(x) = x+1$', color='g')
plt.fill_between(x_vals, f_vals, g_vals, where=(a<=x_vals) & (x_vals<=b), color='orange', alpha=0.5)
plt.axvline(a, color='r', linestyle='--')
plt.axvline(b, color='r', linestyle='--')
plt.title('Área entre as funções $f(x)$ e $g(x)$ entre $x=0$ e $x=2$')
plt.legend()
plt.show()

area_between_functions = abs(sp.integrate((-x**2 + 4*x + 3) - (-x**3 + 7*x**2 - 10*x + 5), (x, a, b)))
print("Área entre as curvas do ponto 0 ao ponto 2:",abs(integral_value_f - integral_value_g), area_between_functions)
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')

f = (x**2)+1
g = x + x**(3/3)

a, b = 0, 2

integral_value = float(integrate(f, (x, a, b)))

f = lambdify(x, f, "numpy")
g = lambdify(x, g, "numpy")

x_vals = np.linspace(-1, 3, 1000)

f_vals = f(x_vals)
g_vals = g(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label=r'$f(x) = x^2+1$')
plt.fill_between(x_vals, 0, f_vals, where=(a<=x_vals) & (x_vals<=b), color='pink', alpha=0.5)
plt.axvline(a, color='r', linestyle='--')
plt.axvline(b, color='r', linestyle='--')
plt.title('a')
plt.legend()
plt.show()
print("Valor calculado da integral de 0 a 2: ",integral_value)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, g_vals, label=r'função pós integração')
plt.legend()
plt.show()
```
