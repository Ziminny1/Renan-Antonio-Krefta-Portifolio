1. $
f(x) = (x^2 - 4)(x^3 - 8x-1)
$

Fazendo a regra de produtos temos que:
$
\frac{dy}{dx}(x^2 - 4)(x^3 - 8x-1)+(x^2 - 4)\frac{dy}{dx}(x^3 - 8x-1)
$

Derivando $(x^2 - 4)$:

$
(2x)(x^3 - 8x-1)+(x^2 - 4)\frac{dy}{dx}(x^3 - 8x-1)
$

Derivando $(x^3 - 8x-1)$

$
(2x)(x^3 - 8x-1)+(x^2 - 4)(3x^2 - 8)
$

Produto de $(2x)(x^3 - 8x)$: 

$
(2x^4 - 16x^2-2x)+(x^2 - 4)(3x^2 - 8)
$

Produto de $(x^2 - 4)(3x^2 - 8)$: 

$
(2x^4 - 16x^2-2x)+(3x^4-8x^2)(-12x^2+32)
$

Simplificando temos:

$
  5x^4-36x^2-2x+32
$

Substituindo $x$ por 5 temos:
$
5*5^4-36*5^2-2*5+32 = 2247
$

O Coeficiente angular da reta tangente no ponto 5 é 2247

2. $
f(x) = \sin(x^3)\cos(x)
$

Fazendo a regra de produtos temos que:

$
\frac{dy}{dx}\cos(x)\sin(x^3)+\cos(x)\frac{dy}{dx}(\sin(x^3))
$

Derivando $\cos(x)$:

$
-\sin(x)\sin(x^3)+\cos(x)\frac{dy}{dx}(\sin(x^3))
$

Derivando $\sin(x^3)$

$
-\sin(x)\sin(x^3)+\cos(x)\cos(x^3)
$
Simplificando chegamos a:

$
  3x^2\cos(x)\cos(x^3)-\sin(x)\sin(x^3)
$

Substituindo $x$ por 5 temos:
$
35^2\cos(5)\cos(5^3)-\sin(5)\sin(5^3)
$

O Coeficiente angular da reta tangente no ponto 5 é 16.167625
